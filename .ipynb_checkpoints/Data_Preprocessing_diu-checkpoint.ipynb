```python
import pandas as pd
import random

valid_shops = []

for shop_id, group in df.groupby('shopid'):

    if len(group) < 1200:
        continue

    counts = group['label'].value_counts()

    # Phải có đủ 3 label
    if not all(label in counts for label in [0, 1, 2]):
        continue

    # Random tổng từ 1000–2000
    target_total = random.randint(1200, 2300)
    per_label = target_total // 3

    sample_parts = []

    for label in [0, 1, 2]:
        part = group[group['label'] == label]

        n = min(len(part), per_label)

        sample_parts.append(part.sample(n=n, random_state=42))

    sample_df = pd.concat(sample_parts)

    # Điều chỉnh lại size
    if len(sample_df) < 1200:
        continue

    if len(sample_df) > 2300:
        sample_df = sample_df.sample(n=2300, random_state=42)

    # Kiểm tra tỷ lệ
    ratio = sample_df['label'].value_counts(normalize=True)

    if any(ratio.get(l, 0) < 0.1 for l in [0, 1, 2]):
        continue

    if ratio.max() > 0.7:
        continue

    valid_shops.append(sample_df)

    # Dừng khi đủ 10 shop
    if len(valid_shops) == 10:
        break

print("Số shop đạt:", len(valid_shops))

# =====================
# Gộp dữ liệu
# =====================
if len(valid_shops) < 10:
    raise ValueError(f"Chỉ có {len(valid_shops)} shop đạt điều kiện!")

final_df = pd.concat(valid_shops, ignore_index=True)

print("Tổng số dòng:", len(final_df))

# =====================
# Lưu file
# =====================
final_df.to_excel('shopee_.xlsx', index=False)

print("Đã lưu 10 shop thành công")
```

```python
import pandas as pd
import os
import random

path = os.path.join('shopee_review.xlsx')
df = pd.read_excel(path)

print('Đọc dữ liệu thành công')

# =====================
# Hàm tạo username
# =====================
def gen_username():
    prefixes = ["sunny", "happy", "best", "quick", "cool", "fresh", "super", "top", "easy", "smart"]
    words = ["shop", "store", "mart", "buy", "market", "home", "vn", "shoppe"]

    suffix = random.randint(1, 99)

    patterns = [
        f"{random.choice(prefixes)}{random.choice(words)}{suffix}",
        f"{random.choice(prefixes)}.{random.choice(words)}{suffix}",
        f"{random.choice(prefixes)}_{random.choice(words)}{suffix}",
        f"{random.choice(['nguyen','tran','le','pham','hoang'])}_{random.choice(['anh','minh','hoa','thao','lan'])}{suffix}",
        f"{random.choice(['nguyen','tran','le','pham','hoang'])}_{random.choice(['van','thi'])}_{random.choice(['minh','hoa','lan'])}"
    ]

    return random.choice(patterns)

# =====================
# Xử lý username trống
# =====================

# Tìm các dòng trống
mask_empty = df['username'].isna() | (df['username'].astype(str).str.strip() == "")

empty_indices = df[mask_empty].index.tolist()

# 👉 Chọn 19 dòng KHÔNG điền
indices_to_skip = set(empty_indices[:19])

# Điền cho các dòng còn lại
for idx in empty_indices:
    if idx not in indices_to_skip:
        df.loc[idx, 'username'] = gen_username()

print(f"Tổng dòng trống: {len(empty_indices)}")
print(f"Đã giữ nguyên trống: {len(indices_to_skip)}")
print(f"Đã điền: {len(empty_indices) - len(indices_to_skip)}")

# =====================
# Lưu file
# =====================
df.to_excel('shopee_review_filled.xlsx', index=False)

print("Đã lưu file thành công")
```

```python
df = df.dropna(subset=["comment", "rating"])

df = df.drop_duplicates(subset=["username", "comment"])

df["time"] = pd.to_datetime(df["time"], unit='s')

df["year"] = df["time"].dt.year
df["month"] = df["time"].dt.month

def generate_labels(n):
    labels = []
    alphabet = string.ascii_uppercase
    
    for i in range(n):
        if i < 26:
            labels.append(alphabet[i])
        else:
            labels.append(alphabet[i//26 - 1] + alphabet[i%26])
    return labels

unique_shops = df["shopid"].unique()
shop_labels = generate_labels(len(unique_shops))
shop_mapping = dict(zip(unique_shops, shop_labels))

df["shopid"] = df["shopid"].map(shop_mapping)
df["shopid"] = "Shop " + df["shopid"]

unique_items = df["itemid"].unique()
item_labels = generate_labels(len(unique_items))
item_mapping = dict(zip(unique_items, item_labels))

df["itemid"] = df["itemid"].map(item_mapping)
df["itemid"] = "Item " + df["itemid"]

df.head(10)
```

```python
def load_dict(path, mode="pair"):
    data = {}

    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split('\t')

            # dạng key-value (teencode)
            if mode == "pair":
                if len(parts) >= 2:
                    key = parts[0]
                    value = parts[1].strip()
                    data[key] = value

            # dạng 1 cột (keyword)
            elif mode == "list":
                data[parts[0]] = True   # dùng như set

    return data
```

```python
teencode_dict = load_dict('dicts/teencode.txt', mode="pair")
keyword_dict  = load_dict('dicts/keyword.txt', mode="list")
```

```python
def convert_teencode(text, teencode_dict):
    words = text.split()
    new_words = []

    for word in words:
        if word in teencode_dict:
            new_words.append(teencode_dict[word])
        else:
            new_words.append(word)

    return " ".join(new_words)
```

```python
keywords = list(keyword_dict.keys())

def contains_keyword(text):
    text = str(text).lower()
    return any(kw in text for kw in keywords)
```

```python
def remove_spam(review):
    """Remove spam and tags like [TB], [QC]"""
    review = str(review)

    # Xóa tag [TB], [QC]
    review = re.sub(r"\[(TB|QC)\]", "", review, flags=re.IGNORECASE)

    # Nếu chứa nội dung spam → bỏ luôn
    spam_patterns = [
        r"http\S*",
        r"www\S*",
        r"\.com",
        r"mua ngay",
        r"liên hệ",
        r"zalo",
        r"fb",
        r"facebook",
        r"\d{9,11}"
    ]

    for pattern in spam_patterns:
        if re.search(pattern, review.lower()):
            return ""

    return review.strip()
```

```python
def to_lowercase(text):
    text = str(text).lower()
    return text
```

```python
def remove_all_tag(sentence):
    """Remove tags added from reading files"""
    return re.sub(r'\n|\t|\r', ' ', sentence)
```

```python
def remove_emoji(text):
    if not isinstance(text, str):
        return ""
    
    return emoji.replace_emoji(text, replace='')
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)
```

```python
def remove_duplicate_letters(text):
    return " ".join(
        re.sub(r'(.)\1+', r'\1', w)
        for w in text.split()
    )
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\sÀ-ỹ]', ' ', text)
```

```python
def remove_nonsense_words(text):
    vowels = "aeiouyàáạảãăắằẳẵặâấầẩẫậèéẹẻẽêếềểễệìíịỉĩòóọỏõôốồổỗộơớờởỡợùúụủũưứừửữự"

    words = text.split()
    clean_words = []

    for w in words:

        # bỏ từ quá dài
        if len(w) > 20:
            continue

        # kiểm tra tỷ lệ nguyên âm
        vowel_count = sum(1 for c in w if c in vowels)
        if len(w) > 6 and vowel_count / len(w) < 0.2:
            continue

        clean_words.append(w)

    return " ".join(clean_words)
```

```python
def clean_text(text):
    text = remove_spam(text)
    text = to_lowercase(text)

    text = remove_all_tag(text)
    text = remove_emoji(text)
    text = remove_numbers(text)

    text = remove_duplicate_letters(text)

    # thay dấu câu bằng khoảng trắng
    text = remove_punctuation(text)

    text = convert_teencode(text, teencode_dict)

    # chuẩn hóa khoảng trắng trước khi tokenize
    text = re.sub(r'\s+', ' ', text).strip()

    # bỏ từ rác
    text = remove_nonsense_words(text)

    return text
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_rating_stats(df):
    rating = df["rating"]

    mean = rating.mean()
    median = rating.median()
    mode = rating.mode()[0]
    std = rating.std()

    q1 = rating.quantile(0.25)
    q3 = rating.quantile(0.75)
    min_val = rating.min()
    max_val = rating.max()

    plt.figure(figsize=(8, 5))

    # Boxplot
    sns.boxplot(x=rating)

    # Vẽ các điểm thống kê
    plt.scatter(mean, 0, color="red", label="Mean")
    plt.scatter(median, 0, color="blue", label="Median")
    plt.scatter(mode, 0, color="green", label="Mode")

    # Vẽ text các giá trị
    plt.text(mean, 0.02, f"Mean={mean:.2f}", ha='center')
    plt.text(median, -0.05, f"Median={median:.2f}", ha='center')
    plt.text(mode, 0.05, f"Mode={mode:.2f}", ha='center')

    # Thêm 5-number summary
    plt.axvline(min_val, linestyle="--", color="gray")
    plt.axvline(q1, linestyle="--", color="orange")
    plt.axvline(median, linestyle="--", color="blue")
    plt.axvline(q3, linestyle="--", color="orange")
    plt.axvline(max_val, linestyle="--", color="gray")

    plt.legend()
    plt.title("Rating Statistics (Mean, Median, Mode, 5-number summary)")
    plt.xlabel("Rating")
    plt.yticks([])

    plt.show()
    
plot_rating_stats(df)
```
