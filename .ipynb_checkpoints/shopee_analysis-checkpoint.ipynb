```python
def load_dict(path, mode="pair"):
    data = {}

    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split('\t')

            # dạng key-value (teencode)
            if mode == "pair":
                if len(parts) >= 2:
                    key = parts[0]
                    value = parts[1].strip()
                    data[key] = value

            # dạng 1 cột (keyword)
            elif mode == "list":
                data[parts[0]] = True   # dùng như set

    return data
```

```python
teencode_dict = load_dict('dicts/teencode.txt', mode="pair")
keyword_dict  = load_dict('dicts/keyword.txt', mode="list")
stopword_dict = load_dict('dicts/vietnamese-stopwords.txt', mode="list")
```

```python
def convert_teencode(text, teencode_dict):
    words = text.split()
    new_words = []

    for word in words:
        if word in teencode_dict:
            new_words.append(teencode_dict[word])
        else:
            new_words.append(word)

    return " ".join(new_words)
```

```python
def to_lowercase(text):
    text = str(text).lower()
    return text
```

```python
def remove_labeled_tags(text):
    text = re.sub(r'[^:\n]+:\s*', '', text)
    return text
```

```python
def remove_all_tag(sentence):
    """Remove tags added from reading files"""
    return re.sub(r'\n|\t|\r', ' ', sentence)
```

```python
def remove_emoji(text):
    if not isinstance(text, str):
        return ""
    
    return emoji.replace_emoji(text, replace='')
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)
```

```python
def remove_duplicate_letters(text):
    return " ".join(
        re.sub(r'(.)\1+', r'\1', w)
        for w in text.split()
    )
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\sÀ-ỹ]', ' ', text)
```

```python
def remove_nonsense_words(text):
    vowels = "aeiouyàáạảãăắằẳẵặâấầẩẫậèéẹẻẽêếềểễệìíịỉĩòóọỏõôốồổỗộơớờởỡợùúụủũưứừửữự"

    words = text.split()
    clean_words = []

    for w in words:

        # bỏ từ quá dài
        if len(w) > 20:
            continue

        # kiểm tra tỷ lệ nguyên âm
        vowel_count = sum(1 for c in w if c in vowels)
        if len(w) > 6 and vowel_count / len(w) < 0.2:
            continue

        clean_words.append(w)

    return " ".join(clean_words)
```

```python
def clean_text(text):
    text = to_lowercase(text)
    text = remove_labeled_tags(text)
    text = remove_all_tag(text)
    text = remove_emoji(text)
    text = remove_numbers(text)

    text = remove_duplicate_letters(text)

    # thay dấu câu bằng khoảng trắng
    text = remove_punctuation(text)

    text = convert_teencode(text, teencode_dict)

    # chuẩn hóa khoảng trắng trước khi tokenize
    text = re.sub(r'\s+', ' ', text).strip()

    # bỏ từ rác
    text = remove_nonsense_words(text)

    return text

df['clean_comment'] = df['comment'].apply(clean_text)
df['comment_length'] = df['clean_comment'].apply(lambda x: len(x.split()))
df[['comment', 'clean_comment', 'comment_length']].head(10)
```

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Tokenize
    text = word_tokenize(text, format="text")

    # 2. Remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopword_dict]

    return " ".join(tokens)
```

```python
df['clean_comment'] = df['clean_comment'].apply(preprocess_text)
df.head(10)
```

### Biểu đồ 7: Top Bigrams theo nhãn (Horizontal Bar)

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_ngrams(corpus, n=2, top_k=10):
    vec = CountVectorizer(
        ngram_range=(n, n),
        min_df=2
    ).fit(corpus)

    bag = vec.transform(corpus)

    words = vec.get_feature_names_out()
    counts = bag.sum(axis=0).A1

    return sorted(zip(words, counts), key=lambda x: x[1], reverse=True)[:top_k]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, group in zip(axes, ['POS', 'NEU', 'NEG']):
    corpus = df[df['label'] == group]['clean_comment'].dropna()

    if len(corpus) < 5:
        ax.set_title(f'{group} (không đủ dữ liệu)')
        continue

    top = get_top_ngrams(corpus, n=2, top_k=10)
    words, counts = zip(*top)

    ax.barh(words[::-1], counts[::-1], color=COLORS[group], edgecolor='white')
    ax.set_title(f'Top Bigrams — {group}', fontsize=13, fontweight='bold', color=COLORS[group])
    ax.set_xlabel('Tần suất', fontsize=10)

fig.suptitle('Top 10 Bigrams theo Nhóm Cảm Xúc', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## CHƯƠNG 6: KIỂM ĐỊNH THỐNG KÊ
### 6.1 Z-test: Tỷ lệ NEG có vượt 30% không?

```python
print('=' * 55)
print('6.1 Z-TEST FOR PROPORTION')
print('H₀: p_NEG ≤ 0.30  |  H₁: p_NEG > 0.30')
print('=' * 55)

n_neg   = (df_clean['label'] == 'NEG').sum()
n_total = len(df_clean)
p_hat   = n_neg / n_total
p0      = 0.30

stat, pval = proportions_ztest(count=n_neg, nobs=n_total, value=p0, alternative='larger')

print(f'  Số review NEG  : {n_neg:,}')
print(f'  Tổng review    : {n_total:,}')
print(f'  p̂ (tỷ lệ NEG) : {p_hat:.4f} ({p_hat*100:.2f}%)')
print(f'  Z-statistic    : {stat:.4f}')
print(f'  p-value        : {pval:.4f}')
print()
```

### 6.2 One-sample t-test: Rating TB có đạt 4 sao không?

```python
print('=' * 55)
print('6.2 ONE-SAMPLE T-TEST')
print('H₀: μ = 4  |  H₁: μ ≠ 4')
print('=' * 55)

t_stat, p_val = ttest_1samp(df_clean['rating'], popmean=4)
mean_rating = df_clean['rating'].mean()

print(f'  Rating trung bình thực tế : {mean_rating:.4f}')
print(f'  t-statistic               : {t_stat:.4f}')
print(f'  p-value (2-tailed)        : {p_val:.4f}')
print()
```

### 6.3 Independent t-test: Rating POS vs NEG

```python
print('=' * 55)
print('6.3 INDEPENDENT T-TEST: POS vs NEG')
print('H₀: μ_POS = μ_NEG  |  H₁: μ_POS ≠ μ_NEG')
print('=' * 55)

pos_ratings = df_clean[df_clean['label'] == 'POS']['rating']
neg_ratings = df_clean[df_clean['label'] == 'NEG']['rating']

t_stat, p_val = ttest_ind(pos_ratings, neg_ratings, equal_var=False)  # Welch's t-test

print(f'  Rating TB nhóm POS : {pos_ratings.mean():.4f}')
print(f'  Rating TB nhóm NEG : {neg_ratings.mean():.4f}')
print(f'  t-statistic        : {t_stat:.4f}')
print(f'  p-value            : {p_val:.6f}')
print()
```

### 6.4 One-way ANOVA: Rating giữa 3 nhóm POS/NEU/NEG

```python
print('=' * 55)
print('6.4 ONE-WAY ANOVA: POS vs NEU vs NEG')
print('H₀: μ_POS = μ_NEU = μ_NEG  |  H₁: Ít nhất 1 nhóm khác')
print('=' * 55)

neu_ratings = df_clean[df_clean['label'] == 'NEU']['rating']
f_stat, p_val = f_oneway(pos_ratings, neu_ratings, neg_ratings)

print(f'  Rating TB POS : {pos_ratings.mean():.4f}')
print(f'  Rating TB NEU : {neu_ratings.mean():.4f}')
print(f'  Rating TB NEG : {neg_ratings.mean():.4f}')
print(f'  F-statistic   : {f_stat:.4f}')
print(f'  p-value       : {p_val:.6f}')
print()
```

### 6.5 ANOVA theo thời gian: Rating có thay đổi theo quý không?

```python
print('=' * 55)
print('6.5 ANOVA THEO QUÝ')
print('H₀: Rating TB không đổi theo quý  |  H₁: Có sự thay đổi')
print('=' * 55)

df_clean['quarter'] = df_clean['datetime'].dt.to_period('Q').astype(str)
quarter_groups = [g['rating'].values for _, g in df_clean.groupby('quarter') if len(g) >= 10]

if len(quarter_groups) >= 2:
    f_stat, p_val = f_oneway(*quarter_groups)
    print(f'  Số quý phân tích : {len(quarter_groups)}')
    print(f'  F-statistic      : {f_stat:.4f}')
    print(f'  p-value          : {p_val:.4f}')
```

## CHƯƠNG 7: XÂY DỰNG & ĐÁNH GIÁ MÔ HÌNH
### 7.1 TF-IDF Vectorization & Train/Test Split

```python
# Loại bỏ dòng rỗng
df_model = df_clean[df_clean['clean_comment'].str.strip() != ''].copy()

X = df_model['clean_comment']
y = df_model['label_num']  # NEG=0, NEU=1, POS=2

# TF-IDF
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2), sublinear_tf=True)
X_tfidf = tfidf.fit_transform(X)

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42, stratify=y
)

print(f'TF-IDF shape     : {X_tfidf.shape}')
print(f'   Train size       : {X_train.shape[0]:,}')
print(f'   Test size        : {X_test.shape[0]:,}')
```

### 7.2 Huấn luyện & So sánh 2 mô hình

```python
label_names = ['NEG', 'NEU', 'POS']
results = {}

# --- Logistic Regression ---
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
results['Logistic Regression'] = {
    'acc'     : accuracy_score(y_test, y_pred_lr),
    'macro_f1': f1_score(y_test, y_pred_lr, average='macro'),
    'y_pred'  : y_pred_lr,
    'model'   : lr
}

# --- Naive Bayes ---
nb = MultinomialNB(alpha=0.5)
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
results['Naive Bayes'] = {
    'acc'     : accuracy_score(y_test, y_pred_nb),
    'macro_f1': f1_score(y_test, y_pred_nb, average='macro'),
    'y_pred'  : y_pred_nb,
    'model'   : nb
}

compare_df = pd.DataFrame({
    'Mô hình'    : list(results.keys()),
    'Accuracy'   : [f"{results[m]['acc']:.4f}" for m in results],
    'Macro F1'   : [f"{results[m]['macro_f1']:.4f}" for m in results],
})
print(compare_df.to_string(index=False))
```

### 7.3 Classification Report chi tiết

```python
for model_name, res in results.items():
    print(classification_report(y_test, res['y_pred'], target_names=label_names))
```

### 7.4 Confusion Matrix

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, (model_name, res) in zip(axes, results.items()):
    cm = confusion_matrix(y_test, res['y_pred'])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names)
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(f'{model_name}\nAcc={res["acc"]:.3f} | Macro F1={res["macro_f1"]:.3f}',
                 fontsize=12, fontweight='bold')

fig.suptitle('Confusion Matrix — So sánh 2 mô hình', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 7.5 Feature Importance (Top từ dự báo mạnh — Logistic Regression)

```python
feature_names = np.array(tfidf.get_feature_names_out())
top_k = 15
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, (ax, group) in enumerate(zip(axes, label_names)):
    coef = lr.coef_[i]
    top_idx = np.argsort(coef)[-top_k:][::-1]
    top_words = feature_names[top_idx]
    top_vals  = coef[top_idx]
    ax.barh(top_words[::-1], top_vals[::-1], color=COLORS[group], edgecolor='white')
    ax.set_title(f'Top từ → {group}', fontsize=13, fontweight='bold', color=COLORS[group])
    ax.set_xlabel('Hệ số (coefficient)', fontsize=10)

fig.suptitle('Feature Importance — Logistic Regression', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
